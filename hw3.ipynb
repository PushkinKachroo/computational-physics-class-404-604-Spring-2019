```python
import numpy as np

def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

def dericd(y,x,dx):
    return (y(x+(dx/2))-y(x-(dx/2)))/dx

def deried(y,x,dx):
    return (4*dericd(y,x,dx/2)-dericd(y,x,dx))/3
```

```python
import numpy as np
x=1.0
dx=0.1
print('forward difference',derifw(np.cos,x,dx))
print('central difference',dericd(np.cos,x,dx))
print('extrapolated difference',deried(np.cos,x,dx))
print('analytical result',-np.sin(x))
```

```python
import numpy as np
import math
import matplotlib.pylab  as p 
x=math.pi/4
logdx=[]
logerr=[]
for n in range(16):
    dx = 10**(n-16)
    logdx.append(math.log(dx,10))
    err = abs(derifw(np.cos,x,dx)+np.sin(x))
    logerr.append(math.log(err,10))
p.plot(logdx,logerr)
p.show()
```
